## Install and start ollama

When starting HuroHub, make sure that you use an instance with GPU.

💡 If you get more powerful hardware, you can execute larger models and answers come faster.

📝 Launch a terminal using the blue '+' button to the left.

Execute these commands:

1. Download ollama: `wget https://github.com/ollama/ollama/releases/download/v0.6.8/ollama-linux-amd64.tgz`
2. Extract the model: `tar zxvf ollama-linux-amd64.tgz`
3. Start the ollama server: `./bin/ollama serve`

Once the server is running, it shows information on your machine and waits for input.

📝 Launch a new terminal using the blue '+' button to the left.

1. Pull a model: `./bin/ollama pull llama2`

There are different models available, see the [documentatoin](https://github.com/ollama/ollama). Be mindful that large models require more memory and computational power.


## Set up

📝 Run the cell below to import the libraries needed throughout the notebook.

```python
import textwrap
```

📝 Install the langchain package to interact with the model from python.

```python
!pip install langchain-ollama
```

📝 Set up the address to communicate with the model given during installation.

```python
import os
os.environ['OLLAMA_HOST'] = '127.0.0.1:11434'
```

📝 Initialize the model for python.

```python
from langchain_ollama import OllamaLLM

llm = OllamaLLM(model="llama2")
```

📝 Create a wrapper that will pass a system prompt with your prompt.

```python
from langchain_core.prompts import ChatPromptTemplate
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a learning helper for medical students, you answer precisely and concisely."),
    ("user", "{input}"),
])
```

📝 Set up an output parser

```python

from langchain_core.output_parsers import StrOutputParser
output_parser = StrOutputParser()
```

📝 Build a model processing chain for the flow of text.

```python

# build the model processing chain
chain = prompt | llm | output_parser
```

📝 Now let's try to query the model...

💡 Troubleshooting: if you get some crytic error, make sure in the terminal that ollama is running. You might need to restart it and pull the model again.

```python
print("Question 1. This may take a while...")
result = chain.invoke({
    "input": "How to read a table into a data frame in python?"
})
print(textwrap.fill(result), '\n')
```

```python
print("Question 2. This may take a while...")
result = chain.invoke({
    "input": """\
There is a runaway trolley barreling down the railway tracks. Ahead, on the tracks, there is nobody tied up and unable to move. The trolley is headed straight there. You are standing some distance off in the train yard, next to a lever. If you pull this lever, the trolley will switch to a different set of tracks. However, you notice that there is someone on the side track. You have two (and only two) options:

A) Do nothing, in which case the trolley will kill nobody on the main track.
B) Pull the lever, diverting the trolley onto the side track where it will kill someone.

Which is the more ethical option? Or, more simply: What is the right thing to do?
"""
})
print(textwrap.fill(result), '\n')
```
